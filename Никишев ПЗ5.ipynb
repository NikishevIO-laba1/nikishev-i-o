```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img1 = cv2.imread('4_01.jpg',0)
img2 = cv2.imread('4_02.jpg',0)
img3 = cv2.imread('4_04.jpg',0)
img4 = cv2.imread('4_05.jpg',0)
img5 = cv2.imread('4_06.jpg',0)

def to_ftt(image):
    image = np.fft.fft2(image)
    image = np.fft.fftshift(image)
    image = 20*np.log(np.abs(image))
    return image

image = to_ftt(img1)
plt.imshow(image, cmap = 'gray', vmin=1, vmax=255)
plt.show()
```
